You are a Data Analyst working for a real estate investment firm. Your job is to analyze house sales data and build a model to predict house prices.

## 1. Importing Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import cross_val_score, train_test_split
%matplotlib inline
```

## 2. Importing the Dataset

```python
# Load the dataset
file_name = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/kc_house_data_NaN.csv'
df = pd.read_csv(file_name)
df.head()
```

## Module 1: Importing Data

```python
# Question 1: Display the data types of each column
print(df.dtypes)
```

## Module 2: Data Wrangling

```python
# Question 2: Drop unnecessary columns and get statistical summary
df.drop(['id', 'Unnamed: 0'], axis=1, inplace=True)
df.describe()
```

```python
# Handle missing values
print("Number of NaN values in bedrooms:", df['bedrooms'].isnull().sum())
print("Number of NaN values in bathrooms:", df['bathrooms'].isnull().sum())

mean_bed = df['bedrooms'].mean()
df['bedrooms'].replace(np.nan, mean_bed, inplace=True)

mean_bath = df['bathrooms'].mean()
df['bathrooms'].replace(np.nan, mean_bath, inplace=True)

print("\nAfter replacement:")
print("NaN bedrooms:", df['bedrooms'].isnull().sum())
print("NaN bathrooms:", df['bathrooms'].isnull().sum())
```

## Module 3: Exploratory Data Analysis

```python
# Question 3: Unique floor values
df['floors'].value_counts().to_frame()
```

```python
# Question 4: Boxplot - Waterfront vs Price
plt.figure(figsize=(8,6))
sns.boxplot(x="waterfront", y="price", data=df)
plt.title("Price Distribution by Waterfront View")
plt.show()
```

```python
# Question 5: Regression plot - sqft_above vs Price
plt.figure(figsize=(8,6))
sns.regplot(x="sqft_above", y="price", data=df, ci=None)
plt.title("sqft_above vs Price")
plt.show()
```

```python
# Correlation with price
df.corr()['price'].sort_values(ascending=False)
```

## Module 4: Model Development

```python
# Question 6: Simple Linear Regression (sqft_living)
X = df[['sqft_living']]
Y = df['price']

lm = LinearRegression()
lm.fit(X, Y)
print("R² Score:", lm.score(X, Y))
```

```python
# Question 7: Multiple Linear Regression
features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement", 
            "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]

X = df[features]
Y = df['price']

lm.fit(X, Y)
print("Multiple Linear Regression R²:", lm.score(X, Y))
```

```python
# Question 8: Pipeline with Scaling + Polynomial + Linear Regression
Input = [('scale', StandardScaler()), 
         ('polynomial', PolynomialFeatures(include_bias=False)), 
         ('model', LinearRegression())]

pipe = Pipeline(Input)
pipe.fit(X, Y)
print("Pipeline R² Score:", pipe.score(X, Y))
```

## Module 5: Model Evaluation and Refinement

```python
# Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)
print("Number of test samples:", x_test.shape[0])
print("Number of training samples:", x_train.shape[0])
```

```python
# Question 9: Ridge Regression
ridge = Ridge(alpha=0.1)
ridge.fit(x_train, y_train)
print("Ridge Regression R² (Test):", ridge.score(x_test, y_test))
```

```python
# Question 10: Second Order Polynomial + Ridge
pr = PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.fit_transform(x_test)

ridge2 = Ridge(alpha=0.1)
ridge2.fit(x_train_pr, y_train)
print("Polynomial Ridge R² (Test):", ridge2.score(x_test_pr, y_test))
```
